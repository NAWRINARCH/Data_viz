```python
import seaborn as sns
#other libraries are matplotlib and plotlib
```

```python
data = sns.load_dataset("tips")
```

```python
data
```

```python
sns.scatterplot(x = "tip", y = "total_bill", data=data, color='red')
```

```python
sns.jointplot(x="tip", y="total_bill", data=data, kind='hex', color ='red')
#color code, color density also histogram shows how the individual data is distributed)
#title = ax.set-title("joint Plot")
#title.set_position ([0.5, 1005])
#sequence mandatory
#joint plot (x-axis and y-axis - continuous)
#['scatter', 'hist', 'hex', 'kde', 'reg', 'resid']
```

```python
data2=data[['tip', 'total_bill']]
```

```python
sns.pairplot(data2)
import matplotlib.pyplot as plt
plt.show()
```

```python
g=sns.pairplot(data,diag_kind ='kde', hue='smoker', palette ='copper')
g.map_upper(sns.scatterplot)
g.map_diag(sns.histplot)
g.map_lower(sns.kdeplot)
g.add_legend()
```

```python
sns.distplot(data['tip'],kde=True, color="yellow")
#kernel density estimation
```

```python
sns.displot(data['tip'],bins=5)
#less the bin size - implicates that each bin is being allocated more data
#thereby increasing each bin width
plt.show()
```

```python
sns.countplot(x="day", data=data, palette ="Purples")
#This function is used to show the counts of observations in each category of a categorical variable.
```

```python
data["smoker"].value_counts()
```

```python
ax=sns.countplot(x='smoker', data=data, palette='Purples')
for p in ax.patches:
    ax.annotate(f'{p.get_height()}',
                (p.get_x() + p.get_width()/2, p.get_height()),
                ha='center',
                va='bottom',
                color='black',
                xytext=(0,-20),
                textcoords='offset points')

#Annotate(annotate meaning adding notes to a text or a diagram) the count plot with value counts
```

```python
#boxplot er box- 25%-75%
#upper fence UF = Q3 +1.5(Q3-Q1)
#lower Fence LF =Q1-1.5(Q3-Q1)
#box at least 1 continuous data mandatory, categorical data thakle compare korbe duta category er
```

```python
sns.boxplot(x="total_bill", y="smoker", data=data, orient="h",
            saturation= 5, palette ="gist_stern_r")

```

```python
sns.boxplot(y="total_bill", x="day", hue='smoker', data=data)
```

```python
sns.violinplot(x="total_bill" , y="day", hue ="smoker", data=data, palette="Reds")
```

```python
#Violin plot - shows the distribution of data, ignore the outliers, upper fence , lower fence, median(q2)
#max width of the violin represents the max spread of the data
#combination of both box plot and kde
#one categorical and 1 continuous
#hue can be added(categorical)
```

```python
sns.barplot(y = "total_bill", x="sex", hue= "smoker", data=data, palette="rainbow")
```

```python
sns.barplot(x="total_bill" , y= "sex", orient="h" , data=data, palette="Blues")
```
